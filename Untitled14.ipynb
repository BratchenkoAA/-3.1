```python
# Задание task_03_01_01.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ

def sgn(x):
    """Вернуть результат сигнум-функции.

    Параметры:
        - x (float): аргумент x.

    Результат:
        float: значение функции.
    """
    if x < 0:
      return -1
    elif x > 0:
      return 1
    else:
      return 0

x = float(input('Введите x: '))
y = float(input('Введите y: '))

sgn_x= sgn(x)
sgn_y= sgn(y)

denominator= sgn_y - (x**0.5)

if denominator != 0:
  z= (sgn_x + (y**2))/ denominator
else:
  z = float('inf')


print("Ответ:", z)

# --------------
# Пример вывода:
#
# Введите x: -9
# Введите y: 0
# Ответ: 0.33
```

```python
# Задание task_03_01_02.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ



def avg(data):
    """Вернуть среднее значение 'data'.

    Параметры:
        data(list): Список чисел, для которых нужно вычеслить среднее значение.

    Результат:
        float: Среднее значение элементов списка
    """
    if not data:
      return 0
    return sum(data) / len(data)


def cleared_data(data):
    """Вернуть список без значений None.

    Параметры:
        data(list): Список, из которого нужно удалить значение None.

    Результат:
        list: Список без значений None.
    """
    return [x for x in data if x is not None]


n = int(input('Кол-во измерений: '))

# Если с клавиатуры вводится прочерк, измерения нет
# (необходимо добавить в список None)
data= []

for i in range(n):
  measurement = input(f'Измерение {i +1}-е: ')
  if measurement== '-':
    data.append(None)
  else:
    data.append(float(measurement))

# Получить очищенный список и среднее значение
cleaned_data= cleared_data(data)
average_value= avg(cleaned_data)



print(f'Средняя температура: {average_value}')
# --------------
# Пример вывода:
#
# Кол-во измерений: 3
# Измерение 1-е: 10
# Измерение 2-е: -
# Измерение 3-е: 20
# Средняя температура: 15.00
```

```python
# Задание task_03_01_03.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ



def is_lucky(num):
    """Является ли 'num' номером счастливого билета?

    Параметры:
        num(int): Номер билета.

    Результат:
        bool: True - да, False - нет.
    """
    even_count= 0
    odd_count= 0
    for digit in str(num):
      if int(digit) %2 == 0:
        even_count+= 1
      else:
        odd_count+= 1
    return even_count == odd_count


def lucky_numbers(a, b):
    """Вернуть список счастливых номеров от 'a' до 'b'.

    Параметры:
        a(int):Начальный номер.
        b(int):Конечный номер.

    Результат:
        list: Список счастливых номеров.
    """
    return [num for num in range(a, b+ 1) if is_lucky(num)]


a = int(input('Первый номер билета: '))
b = int(input('Последний номер билета: '))

# Вывод должен быть осуществлен в строку с одним пробелом-разделителем
lucky_nums= lucky_numbers(a,b)

print(' '.join(map(str, lucky_nums)))
# --------------
# Пример вывода:
#
# Первый номер билета: 10
# Последний номер билета: 25
# 10 12 14 16 18 21 23 25
```

```python
# Задание task_03_01_06.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ



def gcd(first, second):
    """Вернуть НОД целых чисел 'first' и 'second'."""
    while second:
        first, second = second, first % second
    return abs(first)

def lcm(first, second):
    """Вернуть НОК целых чисел 'first' и 'second'."""
    return abs(first * second) // gcd(first, second)


def gcd_nums(nums):
    """Вернуть НОД чисел из списка 'nums'.

    Параметры:
        nums(list): list целых чисел.

    Результат:
       int: НОД чисел из списка.: .
    """
    from functools import reduce
    return reduce(gcd, nums)


def lcm_nums(nums):
    """Вернуть НОК чисел из списка 'nums'.

    Параметры:
        nums(list): Список целых чисел.

    Результат:
        int: НОК чисел из списка.
    """
    from functools import reduce
    return reduce(lcm, nums)


nums = list(map(int, input('Введите числа через пробел: ').split()))

gcd_result = gcd_nums(nums)
lcm_result = lcm_nums(nums)

print(f"НОД = {gcd_result}")
print(f"НОК = {lcm_result}")

# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48
```

```python
# Задание task_03_01_09.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ



LETTERS_EX = {10: "A", 11: "B", 12: "C", 13: "D", 14: "E", 15: "F"}
DIGITS_EX = {"A": 10, "B": 11, "C": 12, "D": 13, "E": 14, "F": 15}


def to_base(number, base):
    """Преобразовать десятичное число 'number' в с.с. 'base'.

    Параметры:
        number (int): десятичное число (N >= 0);
        base (int): система счисления (2 <= base <= 16).

    Результат:
        str: число в с.с. 'base'."""
    if number == 0:
        return '0'

    if base < 2 or base > 16:
        raise ValueError("Основание должно быть в диапазоне от 2 до 16.")

    result = ""
    while number > 0:
        remainder = number % base
        if remainder >= 10:
            result = LETTERS_EX[remainder] + result
        else:
            result = str(remainder) + result
        number //= base

    return result



DIGITS_EX = {"A": 10, "B": 11, "C": 12, "D": 13, "E": 14, "F": 15}
def from_base(number, base):
    """Преобразовать число в с.с. 'base' в десятичное число.

    Параметры:
        number (str): число в с.с. 'base';
        base (int): система счисления (2 <= base <= 16).

    Результат:
        int: число в 10-й с.с."""
    if base < 2 or base > 16:
        raise ValueError("Основание должно быть в диапазоне от 2 до 16.")

    result = 0
    power = 0
    for digit in reversed(number):
        if digit in DIGITS_EX:
            value = DIGITS_EX[digit]
        else:
            value = int(digit)
        if value >= base:
            raise ValueError(f"Цифра '{digit}' не может быть в системе счисления с основанием {base}.")
        result += value * (base ** power)
        power += 1

    return result


print(to_base(255, 16))
print(from_base("FF", 16))
print(to_base(10, 2))
print(from_base("1010", 2))

# --------------
# Пример вывода:
#
# Нет
```

```python
# Задание task_03_01_11.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ



def ceasar(text, shift):
    """Вернуть измененную строку 'text' со сдвигом 'shift'.

    Параметры:
        - text (str): строка;
        - shift (int): свдиг.

    Результат:
        str: измененная строка."""
    # Набор кириллических букв
    letters = [chr(i) for i in range(ord('а'), ord('я') + 1)]
    letters_upper= [chr(i) for i in range(ord('А'), ord('я') + 1)]
    letter_to_index= {letter: index for index, letter in enumerate(letters)}
    letter_to_index_upper= {letter: index for index, letter in enumerate(letters_upper)}
    shifted_text = ''
    for char in text:
      if char in letter_to_index:
        new_index= (letter_to_index[char]+ shift )% len(letters)
        shifted_text += letters[new_index]
      elif char in letter_to_index_upper:
        new_index=(letter_to_index_upper[char]+ shift )% len(letters_upper)
        shifted_text+= letters_upper[new_index]
      else:
        shifted_text+= char
    return shifted_text

text = input('Введите предложение: ')
shift = int(input('Введите сдвиг: '))

encoded = ceasar(text, shift)
decoded = ceasar(encoded, -shift)
print("Зашифрованная строка:", encoded)
print("Расшифрованная строка:", decoded)

# --------------
# Пример вывода:
#
# Введите предложение: ПрограММиРОВание С++
# Введите сдвиг: 4
# Зашифрованная строка: УфтзфдРРмФТЖдсмй Х++
# Расшифрованная строка: ПрограММиРОВание С++
```

```python
# Задание task_03_01_12.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ



# В данной задаче ввод с клавиатуры не нужен.
#
# Используйте пример данных ниже, при необходимости измените его для
# проверки правильности решения



data = [
    {1: 'м', 2: 'м', 3: 'м', 4: 'ж'},
    {1: 'ж', 2: 'м', 3: 'ж', 4: 'ж'},
    {1: 'ж', 2: 'ж', 3: 'ж', 4: 'ж'},
    {1: 'м', 2: 'м', 3: 'м', 4: 'м'},
    {1: None, 2: None, 3: None, 4: None},
    {1: 'м', 2: None, 3: None, 4: 'ж'},
    {1: None, 2: None, 3: None, 4: None},
    {1: 'м', 2: 'м', 3: None, 4: 'м'},
    {1: 'ж', 2: None, 3: None, 4: 'ж'}
]

def vacant_compartments(data):
    """Вернуть список полностью свободных купе. Нумерация купе идет с 1."""
    return [i + 1 for i, compartment in enumerate(data) if all(seat is None for seat in compartment.values())]

def vacant_seats(data, compartments_condition=None, seat_condition=None):
    """Вернуть список свободных мест при соблюдении условий
    'compartments_condition' и 'seat_condition'.
    Нумерация купе и мест идет с 1."""
    result = []
    for i, compartment in enumerate(data):
        if compartments_condition is None or compartments_condition(compartment):
            for j, seat in compartment.items():
                if seat_condition is None or (seat is None and seat_condition(j, seat)):
                    result.append((i + 1, j))
    return result

def is_same_sex_and_vacant(compartment, sex):
    """Вернуть True, если в купе 'compartment' есть свободные места,
    а остальные пассажиры только пола 'sex'."""
    return any(value is not None and value != sex for value in compartment.values()) == False and any(value is None for value in compartment.values())

# список полностью свободных купе
print(vacant_compartments(data))
# список свободных мест в вагоне
print(vacant_seats(data))
# список свободных нижних мест
print(vacant_seats(data, seat_condition=lambda seat, value: seat % 2 != 0))
# список свободных верхних мест
print(vacant_seats(data, seat_condition=lambda seat, value: seat % 2 == 0))
# список свободных мест в купе с исключительно мужской компанией
print(vacant_seats(data, lambda x: is_same_sex_and_vacant(x, "м")))
# список свободных мест в купе с исключительно женской компанией
print(vacant_seats(data, lambda x: is_same_sex_and_vacant(x, "ж")))


# --------------
# Пример вывода:
#
# [5, 7]
# [(5, 1), (5, 2), (5, 3), (5, 4), (6, 2), (6, 3), (7, 1), (7, 2), (7, 3),
#  (7, 4), (8, 3), (9, 2), (9, 3)]
# [(5, 1), (5, 3), (6, 3), (7, 1), (7, 3), (8, 3), (9, 3)]
# [(5, 2), (5, 4), (6, 2), (7, 2), (7, 4), (9, 2)]
# [(8, 3)]
# [(9, 2), (9, 3)]
```

```python
# Задание task_03_01_13.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ



# В данной задаче ввод с клавиатуры не нужен.
#
# Используйте значения 'votes', при необходимости измените его для
# проверки правильности решения

votes = [
    2, 3, -1, 2, 5, 1, 1, 4, 1, 1, 1, 3, 1, 3, 5, -1, 5, 2, 5, 5,
    3, 3, 2, 3, 3, 2, 2, 1, 3, 2, 5, 2, 2, 4, 1, 1, 3, 2, 2, 3, 3,
    3, 1, 4, 2, 1, 4, 2, 3, 3, 3, -1, 5, 3, 1, 4, 5, 1, 1, 3, 3,
    3, -1, 5, 3, 3, 2, 5, 1, 1, 5, -1, 1, 2, 2, 3, -1, 4, 2, 5, 4,
    2, -1, 3, 1, 4, 3, 5, 4, 1, 5, 3, 2, 4, 2, 1, 3, 4, 1, 1, 3, 4
]

parties = {
    1: "Партия №1", 2: "Партия №2", 3: "Партия №3",
    4: "Партия №4", 5: "Партия №5", -1: "Испорчено"
}


def parties_votes(parties, votes):
    """Вернуть информацию о голосах 'votes', отданных за партии 'parties'.
    Испорченные бланки также присутствуют в результате.

    Параметры:
        - parties (dict): информация о партиях (номер голоса: название);
        - votes (list): номера голосов.

    Результат:
        dict: название: кол-во отданных голосов."""

    results = {name: 0 for name in parties.values()}  # Инициализация словаря с нулями
    for vote in votes:
        if vote in parties:
            results[parties[vote]] += 1  # Увеличиваем счетчик для соответствующей партии

    return results

def print_results(votes_for_p):
    """Вывести результаты голосования в формате:

    1. Партия №2 | 1111 | 58.21%
    2. Партия №4 |  999 | 38.14%

    Параметры:
        - votes_for_p (dict): результат функции parties_votes().
    """
    total_votes = sum(votes_for_p.values())  # Общее количество голосов
    sorted_results = sorted(votes_for_p.items(), key=lambda x: x[1], reverse=True)  # Сортировка по количеству голосов

    for index, (party_name, count) in enumerate(sorted_results, start=1):
        percentage = (count / total_votes * 100) if total_votes > 0 else 0  # Вычисление процента
        print(f"{index}. {party_name} | {count:3} | {percentage:.2f}%")

print_results(parties_votes(parties, votes))




# --------------
# Пример вывода:
#
# 1. Партия №3 | 27 | 26.47%
# 2. Партия №1 | 22 | 21.57%
# 3. Партия №2 | 20 | 19.61%
# 4. Партия №5 | 14 | 13.73%
# 5. Партия №4 | 12 | 11.76%
# 6. Испорчено |  7 |  6.86%
```

```python
# Задание task_03_01_14.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ


def split_numbers(*args):
    """Вернуть 2 списка - отрицательных и неотрицательных значений.

    Результат:
        - list: отрицательные значения;
        - list: неотрицательные значения."""

    negative = [num for num in args if num < 0]
    non_negative = [num for num in args if num >= 0]

    return sorted(negative, reverse=True), sorted(non_negative)

# Пример использования
print(split_numbers(1, 4, -5, 0, -33))

# --------------
# Пример вывода:
#
# ([-5, -33], [0, 1, 4])
```

```python
# Задание task_03_01_15.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ


def gdp(c, i, g, ex, im):
    """Вернуть значение ВВП.

Параметры:
        - c: Конечное потребление (int).
        - i: Валовое накопление капитала (инвестиции) (int).
        - g: Государственные расходы (int).
        - ex: Экспорт (int).
        - im: Импорт (int).

    Результат:
        - int: Значение ВВП.
    """
    return c + i + g + ex - im


info_dict = {}
info_list = []


info_dict['c'] = int(input("Конечное потребление: "))
info_dict['i'] = int(input("Валовое накопление капитала: "))
info_dict['g'] = int(input("Государственные расходы: "))
info_dict['ex'] = int(input("Экспорт: "))
info_dict['im'] = int(input("Импорт: "))

info_list = [info_dict['c'], info_dict['i'], info_dict['g'], info_dict['ex'], info_dict['im']]

gdp_value = gdp(*info_list)


print(f"Конечное потребление: {info_dict['c']}")
print(f"Валовое накопление капитала: {info_dict['i']}")
print(f"Государственные расходы: {info_dict['g']}")
print(f"Экспорт: {info_dict['ex']}")
print(f"Импорт: {info_dict['im']}")
print(f"ВВП = {gdp_value}")
# --------------
# Пример вывода:
#
# Конечное потребление: 100
# Валовое накопление капитала: 200
# Государственные расходы: 300
# Экспорт: 400
# Импорт: 500
# ВВП = 500
# ВВП = 500
```

```python
# Задание task_03_01_16.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ



def pow1(value, power):
    """Вернуть 'value' в степени 'power'.

    Параметры:
        - value: число, которое нужно возвести в степень (int или float).
        - power: степень, в которую нужно возвести число (int).

    Результат:
        - int или float: результат возведения 'value' в степень 'power'.
    """
    return value ** power

def pow2(value, power):
    """Вернуть 'value' в степени 'power'. Рекурсивный алгоритм.

    Параметры:
        - value: число, которое нужно возвести в степень (int или float).
        - power: степень, в которую нужно возвести число (int).

    Результат:
        - int или float: результат возведения 'value' в степень 'power'.
    """
    if power == 0:
        return 1  # любое число в степени 0 равно 1
    elif power < 0:
        return 1 / pow2(value, -power)  # для отрицательной степени
    else:
        return value * pow2(value, power - 1)  # рекурсивный вызов

# Примеры использования
print(pow1(5, 3))  # Обычный алгоритм
print(pow2(5, 3))  # Рекурсивный алгоритм

# Пример вывода:
# 125
# 125
```

```python
# Задание task_03_01_17.
#
# Выполнил: Братченко А.А.
# Группа: 241-ЦИБ


def digits_sum(value):
    """Вернуть сумму цифр числа 'value'. Рекурсивная реализация.

    Параметры:
        - value: число, которое нужно возвести в степень (int или float).
        - power: степень, в которую нужно возвести число (int).

    Результат:
- int или float: результат возведения 'value' в степень 'power'.
    """
    value=abs(value)
    if value == 0:
        return 0
    else:
        return value % 10 + digits_sum(value // 10)


def digits_count(value):
    """Вернуть количество цифр числа 'value'. Рекурсивная реализация.

    Параметры:
        - value: число, которое нужно возвести в степень (int или float).
        - power: степень, в которую нужно возвести число (int).

    Результат:
        - int или float: результат возведения 'value' в степень 'power'.
    """
    value=abs(value)

    if value == 0:
      return 1
    elif value < 10:
      return 1
    else:
      return 1 + digits_count(value // 10)


print(digits_sum(12345))
print(digits_count(12345))

# --------------
# Пример вывода:
#
# 15
# 5
```
